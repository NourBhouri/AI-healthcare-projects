```python
 #import libraies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/molecular-biology/promoter-gene-sequences/promoters.data"
```

```python
names = ["Class", "id", "Sequence"]
data_dna = pd.read_csv(url, names = names)
```

```python
# Each column in a Dataframe is called a Series.
classes = data_dna.loc[:,"Class"]
```

```python
classes
```

```python
# generate list of DNA sequence
sequences = list(data_dna.loc[: , "Sequence"])
dataset = {}

# loop hrough sequences and split into individual nucleotides
for i, seq in enumerate(sequences):

  # split into nucleotides, remove ta characters
  nucleotides = list(seq)
  nucleotides = [x for x in nucleotides if x != '\t']

  # append class assignment
  nucleotides.append(classes[i])

  # add to dataset
  dataset[i] = nucleotides

print(dataset[0])
```

```python
# turn dataset into pandas DataFrame
dframe = pd.DataFrame(dataset)
```

```python
# Unfortunately, we can't run machine learning algorithms on the data in 'String' formats. As a result, we need to switch it to numerical data.
# This can easily be accomplished using the pd.get_dummies() function
numerical_df = pd.get_dummies(df)
numerical_df.head()
```

```python
# We don't need both class columns.  Lets drop one then rename the other to simply 'Class'.
df = numerical_df.drop(columns=['Class_-'])

df.rename(columns = {'Class_+': 'Class'}, inplace = True)
df.head()
```

# ***Modeling***

splitting the dataset

```python

# Use the model_selection module to separate training and testing datasets
from sklearn import model_selection

# Create X and Y datasets for training
X = df.drop(['Class'], axis=1)
y = df['Class']

# define seed for reproducibility
seed = 1

# split data into training and testing datasets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.25, random_state=seed)
```

```python
#define the AI models that we want to implement
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

#define scoring method
scoring ='accuracy'

#define the model to train
names = ["Nearest Neighbors", "Gaussian Process","Decision Tree","Random Forest",
         "Neural Net", "AdaBoost","Naive Bayes","SVM Linear","SVM RBF","SVM Sigmoid"]
classifiers =[
    KNeighborsClassifier(n_neighbors=3),
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5),
    RandomForestClassifier(max_depth=5, n_estimators=10,max_features=1),
    MLPClassifier(alpha=1),
    AdaBoostClassifier(),
    GaussianNB(),
    SVC(kernel='linear'),
    SVC(kernel='rbf'),
    SVC(kernel='sigmoid')
]
models =zip(names,classifiers)
```

# ***Evaluation***

```python
#evaluation
from sklearn.metrics import classification_report, accuracy_score
from sklearn.model_selection import KFold, cross_val_score
import numpy as np # Import numpy

results = []
names=[]
accuracy = []
for name,model in models:
    kfold = model_selection.KFold(n_splits = 10)
    cv_results = model_selection.cross_val_score(model, X_train,y_train,cv=kfold,scoring=scoring)
    # Append the mean of cv_results instead of the entire array
    results.append(cv_results.mean())
    names.append(name)
    msg= "%s: %f (%f)" %(name, cv_results.mean(), cv_results.std())

    print(msg)
    model.fit(X_train,y_train)
    predictions= model.predict(X_test)
    print(name)
    print( classification_report(y_test,predictions))
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Boxplot for Cross-Validation Results
sns.boxplot(x=names, y=results, palette="Set2", ax=axes[0])
axes[0].set_title('Model Comparison - Cross Validation Results', fontsize=16, fontweight='bold')
axes[0].set_xlabel('Models', fontsize=14, labelpad=10, fontweight='bold')
axes[0].set_ylabel('Cross-Validation Accuracy', fontsize=14, labelpad=10, fontweight='bold')
# Rotate and align x-axis tick labels
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=45, ha='right') # Changed line
axes[0].grid(True, axis='y', linestyle='--', alpha=0.7)

# Bar plot for Accuracy on Test Set
sns.barplot(x=names, y=accuracy, palette="Set2", ax=axes[1])
axes[1].set_title('Model Comparison - Accuracy on Test Set', fontsize=16, fontweight='bold')
axes[1].set_xlabel('Models', fontsize=14, labelpad=10, fontweight='bold')
axes[1].set_ylabel('Accuracy', fontsize=14, labelpad=10, fontweight='bold')
# Rotate and align x-axis tick labels
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha='right') # Changed line
axes[1].grid(True, axis='y', linestyle='--', alpha=0.7)

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
